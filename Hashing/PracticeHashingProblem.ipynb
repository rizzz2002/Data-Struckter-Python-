```python
class Node:
  def __init__(self, value=None, next = None):
    self.value = value
    self.next = next

class Hashtable:
  def __init__(self, length):
    n = Node()
    self.ht = [n] * length
    self.length = length

  def display(self):
    count = 0
    for i in self.ht:
      temp = i
      print(count, end=" ")
      while temp!=None:
        if temp.next==None:
          print (temp.value)
        else:
          print (temp.value, end="-->")
        temp = temp.next
      count += 1
      print()

  def __hash_function(self, key):
    sum=0
    for i in range (len(key)):
      if i%2==0:
        sum+=int(ord(key[i]))
    return (sum%(len(self.ht)))

  def insert(self, key, value):
    index=self.__hash_function(key)
    if self.ht[index].value==None:
      self.ht[index]=Node((key,value),None)
      return
    else:
      node=self.ht[index]
      if node.next==None:
        self.ht[index]=Node((key,value),node)
        return
      else:
        temp=node
        count=0
        flag=False
        while temp!=None:
          count+=1
          if count==3:
            flag=True
          temp=temp.next
        if flag:
          for y in range (len(self.ht)):
            if self.ht[y].value==None:
              self.ht[y]=Node((key,value),None)
              return
        else:
          self.ht[index]=Node((key,value),node)
          return


  def delete(self, key):
    for y in range (len(self.ht)):
      if self.ht[y]!=None:
        temp=self.ht[y]
        if temp.next==None:
          if temp.value[0]==key:
            self.ht=Node()
            return
        else:
          while temp.next!=None:
            if temp.next.value[0]==key:
              if temp.next.next.value!=None:
                temp.next=temp.next.next
                return
              else:
                temp.next=Node()
                return
          temp=temp.next


ht = Hashtable(3)


ht.insert("ibanez", 500)
ht.insert("gibson", 600)
ht.insert("yamaha", 400)
ht.insert("epiphone", 350)
ht.insert("fender", 550)
ht.insert("squier", 350)


ht.display()
print("===========================================================")
print("===========================================================")
ht.delete("gibson")
ht.display()
```
